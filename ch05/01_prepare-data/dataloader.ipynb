```python
from pathlib import Path
import requests
```

```python
wiki_titles = [
    '1980年冬季奥林匹克运动会',
    '1980年夏季奥林匹克运动会',
    '1984年冬季奥林匹克运动会',
    '1984年夏季奥林匹克运动会',
    '1988年冬季奥林匹克运动会',
    '1988年夏季奥林匹克运动会',
    '1992年冬季奥林匹克运动会',
    '1992年夏季奥林匹克运动会',
    '1994年冬季奥林匹克运动会',
    '1996年夏季奥林匹克运动会',
    '1998年冬季奥林匹克运动会',    
    '2000年夏季奥林匹克运动会',
    '2002年冬季奥林匹克运动会',
    '2004年夏季奥林匹克运动会',
    '2006年冬季奥林匹克运动会',
    '2008年夏季奥林匹克运动会',
    '2010年冬季奥林匹克运动会',
    '2012年夏季奥林匹克运动会',
    '2014年冬季奥林匹克运动会',
    '2016年夏季奧林匹克運動會',
    '2018年冬季奥林匹克运动会',
    '2020年夏季奥林匹克运动会',
    '2022年冬季奥林匹克运动会',
    "2024年夏季奥林匹克运动会",
]

data_path = Path("data")

for title in wiki_titles:
    response = requests.get(f"https://zh.wikipedia.org/w/api.php", params={
            "action": "query",
        "format": "json",
        "prop": "extracts",
        "titles": title,
        "explaintext": True,
    }).json()
    page = next(iter(response["query"]["pages"].values()))
    wiki_text = page["extract"]
    
    if not data_path.exists():
        data_path.mkdir(exist_ok=True)
    
    with open(data_path / f"{title}.txt", "w", encoding="utf-8") as f:
        f.write(wiki_text) 
```
